### **CASE STUDY - SCHIZOPHRENIA**

**SCHIZOPHERNIA** - A type of mental illness that affects how a person thinks, feels, and behaves. It involves hallucinations, delusions, disorganized thinking, social withdrawal, and many more.

*Dataset credits: https://www.kaggle.com/datasets/asinow/schizohealth-dataset*

The dataset is a comprehensive collection of **demographic, clinical, and psychosocial information** about schizophrenia patients. It serves as a valuable resource for researchers, doctors, and data scientists. The dataset includes **patient's diagnosis status, symptom scores, treatment history, and social factors**.

**PROBLEM STATEMENT**

Given the significant rise in the number of schizophrenia cases, we aim to understand how **demographic, clinical, and psychosocial information** can trigger the symptoms in a person, identify the key responsible factors, and understand the diagnosis based on symptom severity. By combining these approaches, we hope to gain a comprehensive understanding of schizophrenia and contribute to improved early detection and intervention strategies.

**Questions**
- What demographic, clinical, and social factors are most strongly associated with a schizophrenia diagnosis and its severity?
- What are the most significant risk factors for schizophrenia, considering demographic, clinical, and psychosocial variables?
- Are there distinct subgroups of individuals with schizophrenia based on their clinical profiles and risk factor combinations?
- How do factors like disease duration, hospitalization history, family history, and substance use influence diagnosis, symptom burden, and overall functioning?
- What insights can we gain from this analysis to inform early interventions and improve outcomes for individuals with schizophrenia?

**DEFINING THE COLUMNS & THEIR DATATYPES**

- **Patient_ID (Integer)**: Unique identifier assigned to each patient.

- **Age (Integer)**: Patient's age (between 18 and 80).

- **Gender (Categorical)**: 0: Female, 1: Male.

- **Education_Level (Categorical)**: 1: Primary, 2: Middle School, 3: High School, 4: University, 5: Postgraduate.

- **Marital_Status (Categorical)**: 0: Single, 1: Married, 2: Divorced, 3: Widowed.

- **Occupation (Categorical)**: 0: Unemployed, 1: Employed, 2: Retired, 3: Student.

- **Income_Level (Categorical)**: 0: Low, 1: Medium, 2: High.

- **Living_Area (Categorical)**: 0: Rural, 1: Urban.

- **Diagnosis (Categorical)**: 0: Not schizophrenic, 1: Schizophrenic.

- **Disease_Duration (Float)**: Duration of illness for schizophrenia patients (1-40 years).

- **Hospitalizations (Integer)**: Number of hospital admissions (ranges from 0 to 10 for schizophrenia patients).

- **Family_History (Categorical)**: 0: No, 1: Yes (Genetic predisposition).

- **Substance_Use (Categorical)**: 0: No, 1: Yes (Tobacco, alcohol, or other substances).

- **Suicide_Attempt (Categorical)**: 0: No, 1: Yes.

- **Positive_Symptom_Score (Integer)**: Ranges from 0 to 100 (Higher scores indicate more symptoms).

- **Negative_Symptom_Score (Integer)**: Ranges from 0 to 100 (Higher scores indicate more negative symptoms).

- **GAF_Score (Integer)**: Global Assessment of Functioning (Ranges from 0 to 100, lower scores indicate worse functioning).

- **Social_Support (Categorical)**: 0: Low, 1: Medium, 2: High.

- **Stress_Factors (Categorical)**: 0: Low, 1: Medium, 2: High.

- **Medication_Adherence (Categorical)**: 0: Poor, 1: Moderate, 2: Good.

**RENAMING THE COLUMNS AND TRANSFORMING THEIR DTYPES TO DESIRED FORMAT**

```python
df = df.rename(columns={'Hasta_ID':'Patient_ID'})
df = df.rename(columns={'Yaş':'Age'})
df = df.rename(columns={'Cinsiyet':'Gender'})
df['Gender'] = df['Gender'].astype('category')

df = df.rename(columns={'Eğitim_Seviyesi':'Education_Level'})
df['Education_Level'] = df['Education_Level'].astype('category')

df = df.rename(columns={'Medeni_Durum':'Marital_Status'})
df['Marital_Status'] = df['Marital_Status'].astype('category')

df = df.rename(columns={'Meslek':'Occupation'})
df['Occupation'] = df['Occupation'].astype('category')

df = df.rename(columns={'Gelir_Düzeyi':'Income_Level'})
df['Income_Level'] = df['Income_Level'].astype('category')

df = df.rename(columns={'Yaşadığı_Yer':'Living_Area'})
df['Living_Area'] = df['Living_Area'].astype('category')

df = df.rename(columns={'Tanı':'Diagnosis'})
df['Diagnosis'] = df['Diagnosis'].astype('category')

df = df.rename(columns={'Hastalık_Süresi':'Disease_Duration'})
df['Disease_Duration'] = df['Disease_Duration'].astype('float64')

df = df.rename(columns={'Hastaneye_Yatış_Sayısı':'Hospitalizations'})
df = df.rename(columns={'Ailede_Şizofreni_Öyküsü':'Family_History'})
df['Family_History'] = df['Family_History'].astype('category')

df = df.rename(columns={'Madde_Kullanımı':'Substance_Use'})
df['Substance_Use'] = df['Substance_Use'].astype('category')

df = df.rename(columns={'İntihar_Girişimi':'Suicide_Attempt'})
df['Suicide_Attempt'] = df['Suicide_Attempt'].astype('category')

df = df.rename(columns={'Pozitif_Semptom_Skoru':'Positive_Symptom_Score'})
df = df.rename(columns={'Negatif_Semptom_Skoru':'Negative_Symptom_Score'})
df = df.rename(columns={'GAF_Skoru':'GAF_Score'})
df = df.rename(columns={'Sosyal_Destek':'Social_Support'})
df['Social_Support'] = df['Social_Support'].astype('category')

df = df.rename(columns={'Stres_Faktörleri':'Stress_Factors'})
df['Stress_Factors'] = df['Stress_Factors'].astype('category')

df = df.rename(columns={'İlaç_Uyumu':'Medication_Adherence'})
df['Medication_Adherence'] = df['Medication_Adherence'].astype('category')
```

```python
df['Gender'] = df['Gender'].replace({0: 'Female', 1: 'Male'})
df['Education_Level'] = df['Education_Level'].replace({1: 'Primary', 2: 'Middle School', 3: 'High School', 4: 'University', 5: 'Postgraduate'})
df['Marital_Status'] = df['Marital_Status'].replace({0: 'Single', 1: 'Married', 2: 'Divorced', 3: 'Widowed'})
df['Occupation'] = df['Occupation'].replace({0: 'Unemployed', 1: 'Employed', 2: 'Retired', 3: 'Student'})
df['Income_Level'] = df['Income_Level'].replace({0: 'Low', 1: 'Medium', 2: 'High'})
df['Living_Area'] = df['Living_Area'].replace({0: 'Rural', 1: 'Urban'})
df['Diagnosis'] = df['Diagnosis'].replace({0: 'Not Schizophrenic', 1: 'Schizophrenic'})
df['Family_History'] = df['Family_History'].replace({0: 'No', 1: 'Yes'})
df['Substance_Use'] = df['Substance_Use'].replace({0: 'No', 1: 'Yes'})
df['Suicide_Attempt'] = df['Suicide_Attempt'].replace({0: 'No', 1: 'Yes'})
df['Social_Support'] = df['Social_Support'].replace({0: 'Low', 1: 'Medium', 2: 'High'})
df['Stress_Factors'] = df['Stress_Factors'].replace({0: 'Low', 1: 'Medium', 2: 'High'})
df['Medication_Adherence'] = df['Medication_Adherence'].replace({0: 'Poor', 1: 'Moderate', 2: 'Good'})
```

**CHECK THE SKEWNESS AND KURTOSIS OF THE DISTRIBUTION**

- (-0.5, 0.5) — low or approximately symmetric.
- (-1, -0.5) U (0.5, 1) — moderately skewed.
- Beyond -1 and 1 — Highly skewed.

```python
for col in num_col:
    print(f"{col} skewness is : {df[col].skew()}")
```

- Mesokurtic distribution (kurtosis = 3, excess kurtosis = 0): perfect normal distribution or very close to it.
- Leptokurtic distribution (kurtosis > 3, excess kurtosis > 0): sharp peak, heavy tails
- Platykurtic distribution (kurtosis < 3, excess kurtosis < 0): flat peak, light tails

```python
for col in num_col:
    print(f"{col} kurtosis is : {df[col].kurtosis()}")
```

**Conclusion**  

This analysis explored the **demographic, clinical, and psychosocial factors** associated with schizophrenia diagnosis and symptom severity. The results highlight several **notable trends and risk factors**:  

1. **Demographic Trends & Risk Factors**  
   - Schizophrenia cases show a **bimodal age distribution**, with peaks in the **30–40** and **60–70** age groups.  
   - The condition is more prevalent in **low-income, rural populations** and those with a **family history of schizophrenia**.  
   - Higher education levels (especially **postgraduates**) are more frequently diagnosed, which may reflect **greater healthcare access and diagnosis rates** rather than actual risk.  

2. **Clinical Factors & Symptom Patterns**  
   - **Positive and Negative Symptom Scores** are **highly correlated**, indicating shared underlying mechanisms of symptom severity. This correlation suggests that **removing one variable** in predictive modeling could help prevent multicollinearity.  
   - The **Global Assessment of Functioning (GAF) Score** follows a **left-skewed distribution**, suggesting that most individuals experience significant functional impairment.  

3. **Social & Behavioral Factors**  
   - **Marital status** influences risk: **Married, divorced, and widowed individuals** show higher schizophrenia prevalence than single individuals, which could be linked to **stressors in relationships or caregiving burdens**.  
   - **Employment status** reveals an unexpected trend: **Employed individuals** show higher rates, possibly due to workplace stress or increased medical reporting among those with healthcare access.  
   - **Medical adherence** is a critical factor: **Poor medication adherence** is associated with increased schizophrenia severity, highlighting the need for **better treatment compliance strategies**.  
   - **Contrary to expectations, stress and social support did not play a significant role**, suggesting that biological and clinical factors may be more dominant predictors.  

**Implications & Recommendations**  
These findings underscore the **importance of early screening**, particularly for individuals with **family history, low income, rural backgrounds, and poor medication adherence**. Given the **strong correlation between symptom scores**, future predictive models should consider feature selection techniques to avoid multicollinearity. Additionally, **further research** into the **role of employment stress and marital status** in schizophrenia risk could provide deeper insights.  

**Limitations & Future Research**  
While this analysis provides valuable insights, certain **limitations** should be acknowledged:  
- The **skewed distributions** of certain clinical variables suggest that further **statistical transformations** or non-parametric methods may improve analysis.  
- The **bivariate nature** of this analysis does not account for **interactions between multiple risk factors**. Future work should explore **machine learning models** or **regression techniques** to uncover more complex relationships.  
- The **lack of significant association** between social support and schizophrenia risk warrants further investigation, as it contradicts some prior research.  

**Final Thought**  
By understanding these **demographic, clinical, and social factors**, **healthcare practitioners can design better intervention strategies** to identify high-risk individuals early and improve **long-term patient outcomes**.
